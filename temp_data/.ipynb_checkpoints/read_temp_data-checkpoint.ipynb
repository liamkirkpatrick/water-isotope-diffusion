# Script to read in JM's data

Note: save this to the same folder 

```python
import pandas as pd
import matplotlib.pyplot as plt


# set file
file = open('data_proc.pkl', 'rb')

# read data
data = pd.read_pickle(file)
alhic1901 = data['ALHIC1901_22_30min']
depth = alhic1901[0]
temp = alhic1901[3]

# Plot
fig,axs = plt.subplots()
axs.plot(temp,depth)
axs.set_ylim([150,0])
axs.set_xlabel('Temperature (C)')
axs.set_ylabel('Depth (m)')
```

```python
# Asked chat GPT to do this - let's see how it does!

# Create a DataFrame using a dictionary
df = pd.DataFrame({
    'depth': depth,
    'temperature': temp
})

# Save the DataFrame to a CSV file
csv_filename = 'depth_temperature_data.csv'
df.to_csv(csv_filename, index=False)  # Setting index=False to avoid adding an unwanted index column in the CSV

# saves the CSV to the same folder where you have your script. You can then open it in excel!
```
